```python
# Ques - 02
# There is a simulation named robot_motion.mp4
import numpy as np
import matplotlib.pyplot as plt

# Unicycle robot model parameters
v_const = 1.0

# Time parameters
T = 10.0  # total simulation time
dt = 0.1  # time step
num_steps = int(T / dt)

# Control input
def control_input(t):
    if 0 <= t <= 10:
        v = 1.0
        if 0.5 <= t <= 1.5 or 2 <= t <= 3 or 4 <= t <= 5 or 6 <= t <= 7 or 8 <= t <= 9:
            omega = 3.0 if t % 2 == 0 else -3.0
        else:
            omega = 0.0
        return v, omega
    else:
        return 0.0, 0.0

# Unicycle robot model differential equations
def unicycle_model(t, x, y, q):
    v, omega = control_input(t)
    dx_dt = v * np.cos(q)
    dy_dt = v * np.sin(q)
    dq_dt = omega
    return dx_dt, dy_dt, dq_dt

# Euler's method
def euler_method(t0, x0, y0, q0, dt, num_steps):
    t_values = [t0]
    x_values = [x0]
    y_values = [y0]
    q_values = [q0]

    for _ in range(num_steps):
        t, x, y, q = t_values[-1], x_values[-1], y_values[-1], q_values[-1]
        dx, dy, dq = unicycle_model(t, x, y, q)
        x = x + dt * dx
        y = y + dt * dy
        q = q + dt * dq
        t = t + dt

        t_values.append(t)
        x_values.append(x)
        y_values.append(y)
        q_values.append(q)

    return t_values, x_values, y_values, q_values

# Initial conditions
t0, x0, y0, q0 = 0.0, 0.0, 0.0, 1.0

# Run simulation
t_values, x_values, y_values, q_values = euler_method(t0, x0, y0, q0, dt, num_steps)

# Plotting
plt.figure(figsize=(12, 8))

# Plot x vs y
plt.subplot(2, 2, 1)
plt.plot(x_values, y_values, label="Robot Path")
plt.title("x vs y")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Plot x vs t
plt.subplot(2, 2, 2)
plt.plot(t_values, x_values, label="x")
plt.title("x vs t")
plt.xlabel("t")
plt.ylabel("x")
plt.grid(True)
plt.legend()

# Plot y vs t
plt.subplot(2, 2, 3)
plt.plot(t_values, y_values, label="y", color="orange")
plt.title("y vs t")
plt.xlabel("t")
plt.ylabel("y")
plt.grid(True)
plt.legend()

# Plot q vs t
plt.subplot(2, 2, 4)
plt.plot(t_values, q_values, label="q", color="green")
plt.title("q vs t")
plt.xlabel("t")
plt.ylabel("q")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Question B
from matplotlib.animation import FuncAnimation
# Function to update plot in animation
def update_plot(frame, x_values, y_values, line):
    line.set_data(x_values[:frame], y_values[:frame])
    return line,

# Run simulation
t_values, x_values, y_values, q_values = euler_method(t0, x0, y0, q0, dt, num_steps)

# Create an animation
fig, ax = plt.subplots()
line, = ax.plot([], [], label="Robot Path")
ax.set_xlim(min(x_values), max(x_values))
ax.set_ylim(min(y_values), max(y_values))
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Robot Motion")

def init():
    line.set_data([], [])
    return line,

# Number of frames for the animation
num_frames = len(t_values)

# Animation
animation = FuncAnimation(
    fig,
    update_plot,
    frames=num_frames,
    init_func=init,
    fargs=(x_values, y_values, line),
    interval=100,  # Adjust the interval as needed
    blit=True,
)

# Save each frame as an image (PNG)
for i in range(num_frames):
    plt.figure()
    plt.plot(x_values[:i+1], y_values[:i+1], label="Robot Path")
    plt.title("Robot Motion")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim(min(x_values), max(x_values))
    plt.ylim(min(y_values), max(y_values))
    plt.legend()
    plt.savefig(f"frame_{i:03d}.png")
    plt.close()

# To create a video using FFmpeg, run the following command in the terminal:
!ffmpeg -r 10 -i frame_%03d.png -vcodec libx264 -pix_fmt yuv420p robot_motion.mp4

# Display the animation
plt.show()
```

```python
# Ques - 03
import numpy as np
import matplotlib.pyplot as plt

# Define the time step and time range
Dt = 0.1
t = np.arange(0, 10 + Dt, Dt)

# Initialize arrays to store the state variables
x = np.zeros(len(t))
y = np.zeros(len(t))
q = np.zeros(len(t))

# Constants
r = 0.1  # Radius of wheels
L = 1.0  # Distance between wheels

# Initial conditions
x[0] = 0
y[0] = 0
q[0] = 1

# Define the control input functions based on the given conditions
def control_input_u1(t):
    if 4 <= t % 10 <= 6:
        return 12
    elif 6 <= t % 10 <= 8:
        return 12
    else:
        return 1

def control_input_u2(t):
    if 0.5 <= t % 10 <= 1.5:
        return 12
    elif 2 <= t % 10 <= 4:
        return 12
    else:
        return 1

# Simulate the system using Euler's method
for i in range(1, len(t)):
    u1 = control_input_u1(t[i-1])
    u2 = control_input_u2(t[i-1])
    x[i] = x[i-1] + Dt * (r / 2) * (u1 + u2) * np.cos(q[i-1])
    y[i] = y[i-1] + Dt * (r / 2) * (u1 + u2) * np.sin(q[i-1])
    q[i] = q[i-1] + Dt * (r / L) * (u1 - u2)

# Plotting the results
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(y, x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('x vs y')

plt.subplot(2, 2, 2)
plt.plot(t, x)
plt.xlabel('t')
plt.ylabel('x')
plt.title('x vs t')

plt.subplot(2, 2, 3)
plt.plot(t, y)
plt.xlabel('t')
plt.ylabel('y')
plt.title('y vs t')

plt.subplot(2, 2, 4)
plt.plot(t, q)
plt.xlabel('t')
plt.ylabel('q')
plt.title('q vs t')

plt.tight_layout()
plt.show()
```
